```python
import pandas as pd

# Load the dataset
df = pd.read_csv('cleve.csv')

# Identify the outlier values based on IQR method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)

# Remove the outlier rows from the dataset
clean_df = df[~outliers]

# Save the cleaned dataset to a new file
clean_df.to_csv('cleve_cleaned.csv', index=False)

dataset=clean_df
# Importing the dataset

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, 13].values
```

```python
#handling missing data

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')


imputer=imputer.fit(X[:,11:13])
X[:,11:13]=imputer.transform(X[:,11:13])

```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# define the machine learning models to be used
svm = SVC(probability=True)
dt = DecisionTreeClassifier()
nn = MLPClassifier()
knn = KNeighborsClassifier()
nb = GaussianNB()

```

```python
# create a list of tuples containing the name and model object for each base model
import random
import warnings
from sklearn.exceptions import ConvergenceWarning

base_models = [('svm', svm), ('dt', dt), ('nn', nn), ('knn', knn)]
from sklearn.metrics import classification_report
i=0
accuracy=0.001

while i<100:
    # ignore convergence warnings
    warnings.filterwarnings("ignore", category=ConvergenceWarning)
    random.shuffle(base_models)
    rf = RandomForestClassifier(n_estimators=100, random_state=1)

    # define the stacking classifier object
    stack = StackingClassifier(estimators=base_models, final_estimator=rf, cv=5)

    # fit the stacking classifier to the training data
    stack.fit(X_train, y_train)

    # make predictions using the stacking classifier
    y_pred = stack.predict(X_test)

    # evaluate the accuracy of the predictions
    temp = accuracy_score(y_test, y_pred)
#     print(temp)
    
    if temp>accuracy:
        accuracy=temp
        classification=classification_report(y_test, y_pred)
        
    i=i+1
print(classification)
print (accuracy)
```
